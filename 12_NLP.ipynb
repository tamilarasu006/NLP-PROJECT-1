```python
# Breadth First Search (BFS) Program

from collections import deque

# Graph represented using dictionary
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

def bfs(start):
    visited = []        # List to store visited nodes
    queue = deque()      # Queue for BFS

    queue.append(start)  # Insert start node into queue

    while queue:
        node = queue.popleft()   # Remove front element
        if node not in visited:
            print(node, end=" ")
            visited.append(node)

            # Add unvisited adjacent nodes to queue
            for neighbour in graph[node]:
                if neighbour not in visited:
                    queue.append(neighbour)

# Main Program
start_node = 'A'
print("BFS Traversal:")
bfs(start_node)
```

```python
# Depth First Search (DFS) Program

# Graph represented using dictionary
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

visited = []   # List to store visited nodes

def dfs(node):
    if node not in visited:
        print(node, end=" ")
        visited.append(node)

        # Visit all adjacent nodes recursively
        for neighbour in graph[node]:
            dfs(neighbour)

# Main Program
start_node = 'A'
print("DFS Traversal:")
dfs(start_node)
```
